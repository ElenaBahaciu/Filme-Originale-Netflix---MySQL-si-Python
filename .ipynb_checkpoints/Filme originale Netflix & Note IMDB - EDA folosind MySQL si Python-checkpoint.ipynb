# Despre baza de date

Baza de date consta intr-un tabel ce include toate filmele originale Netflix produse pana la 1 Iunie 2021. De asemenea sunt prezente si toate documentarele realizate de Netflix si toate special-urile. Majoritatea filmelor au peste 1000 review-uri pe IMDB.

Baza de date poate fi descarcata de pe [Kaggle](https://www.kaggle.com/datasets/luiscorter/netflix-original-films-imdb-scores). 

Continutul tabelului:

- Title: numele filmului
- Genre: genul sau genurile filmului 
- Premiere:  data de lansare a filmului
- Runtime: durata in minute
- IMDB scores : notele de pe IMDB pana la data de 01/06/21)
- Language: limba/limbile in care sunt disponibile filmele (pana la data de  01/06/21)

```python
#Top 15 productii originale Netflix (ordonate in functie de nota de pe IMDB)
query3='SELECT Title, IMDB_score, Genre, Language\
    FROM netflix_originals\
        ORDER BY IMDB_score DESC\
        LIMIT 15;'
```

```python
#Top 10 cele mai bune documentare Netflix
query4="SELECT Title, Premiere, IMDB_Score, Language\
    FROM netflix_originals\
        WHERE Genre='Documentary'\
        ORDER BY IMDB_Score DESC\
        LIMIT 10;"
```

```python
#Care este in medie nota unui documentar?

query5="SELECT Genre, ROUND(AVG(IMDB_score),2) AS 'Average score'\
    FROM netflix_originals\
    GROUP BY Genre\
    HAVING Genre='Documentary'"
```

```python
q5=pd.read_sql_query(query5,connection)
q5
```

```python
#Cele mai bune 10 productii Netflix din anul 2021
query6="SELECT Title, Genre, IMDB_score, DATE_FORMAT(Premiere, '%M %Y') AS 'Premiere Date'\
    FROM netflix_originals\
        WHERE Premiere > '2021-01-01'\
        ORDER BY IMDB_score DESC\
        limit 10;"
```

```python
q6=pd.read_sql_query(query6,connection)
q6
```

```python
#Cate filme din fiecare categorie sunt in tabel
query7='SELECT Genre, COUNT(*) AS Count\
    FROM netflix_originals\
        GROUP BY Genre\
        ORDER BY Count DESC;'
```

```python
q7=pd.read_sql_query(query7,connection)
q6
```

```python
#Tabelul de mai jos indica, pentru fiecare an in parte, numarul de productii aparute si media notelor de pe IMDB 
query8="SELECT\
 Year(Premiere) AS Year,\
 COUNT(*) AS 'Number of Movies',\
 ROUND(AVG(IMDB_score),1) AS 'Average Score'\
    FROM netflix_originals\
    GROUP BY Year(Premiere)\
    ORDER BY Year(Premiere) DESC;"
```

```python
q8=pd.read_sql_query(query7,connection)
q8
```

```python
#Cate filme cu o nota mai mare de 8 sunt in baza de date?
query9="SELECT COUNT(*) AS 'Number of Good Movies'\
    FROM netflix_originals\
        WHERE IMDB_score>8;"
```

```python
#Care este in medie durata a unui film?
query12="SELECT AVG(Runtime) AS 'Average Runtime'\
    FROM netflix_originals;"
```

```python
q12=pd.read_sql_query(query12,connection)
q12
```

```python
#Afiseaza toate productiile care au o durata mai mare decat media
query13='SELECT Title, Genre, Runtime\
    FROM netflix_originals\
        WHERE Runtime>\
            (SELECT AVG(Runtime) FROM netflix_originals);'
```

```python
q13=pd.read_sql_query(query13,connection)
q13
```

```python
#In tabelul de mai jos filmele sunt clasificate in 3 categorii: Feature film, Short Film si Interview, in functie de durata fieacrul film.
#Pentru fiecare categorie se calculeaza o serie de functii agregate

query13="SELECT \
    CASE\
        WHEN Genre like '%interview' THEN 'Interview'\
        WHEN Runtime <=40 THEN 'Short Film'\
        WHEN Runtime >40 THEN 'Feature Film'\
    END AS 'Film Type',\
    COUNT(*) AS 'Number of Movies',\
    ROUND(AVG(Runtime),1) AS 'Average Runtime',\
    ROUND(AVG(IMDB_score),1) AS 'Average IMDB score',\
    ROUND(MAX(IMDB_score),1) AS 'Highest IMDB score',\
    ROUND(MIN(IMDB_score),1) AS 'Lowest IMDB score'\
    FROM netflix_originals\
    GROUP BY CASE\
        WHEN Genre like '%interview' THEN 'Interview'\
        WHEN Runtime <=40 THEN 'Short Film'\
        WHEN Runtime >40 THEN 'Feature Film'\
    END ;"
```

```python
q14=pd.read_sql_query(query14,connection)
q14
```

**Crearea unei histograme care sa reprezinte distributie notelor de pe IMDB a filmelor din tabel**

```python
plt.figure(figsize=(15,7), dpi=80)
sns.histplot(data=netflix,x='IMDB Score',bins=30,kde=True)
plt.title('Distributie notelor de pe IMDB')

plt.show()
```

**Crearea unei histograme care sa reprezinte distributie duratei filmelor**

```python
plt.figure(figsize=(15,7), dpi=80)
sns.histplot(data=netflix,x='Runtime',bins=30,kde=True)
plt.title('Distributie duratei filmelor')

plt.show()
```

**Crearea unei joint plot intre notele si durata unui film, impartind filmele in 2 categori 'Short Film' si 'Feature Film'**

```python
#Adaug o noua coloana care sa imparta filmele in 'Short Film' si 'Feature Film' in functie de durata filmului.
#Productiile de pana in 40 minute vor fi incadrate la 'Short film', iar cele mai lungi de 40 minute vor fi condiderate 'Feature Film'.

def film_category(x):
    if x < 40:
        return 'Short Film'
    elif x > 40:
        return 'Feature Film'

```

```python
netflix['Category']=netflix['Runtime'].apply(lambda x: film_category(x))
```

```python
netflix
```

```python
#Joint plot intre notele de pe IMDB si durata filmului
plt.figure(figsize=(15,7), dpi=80)
sns.jointplot(data=netflix,x='IMDB Score',y='Runtime', hue='Category')
```

```python
#Recreerea histogramei notelor de pe IMDB cu hue='Category'
plt.figure(figsize=(15,7), dpi=80)
sns.histplot(data=netflix,x='IMDB Score',bins=30,kde=True, hue='Category')
plt.title('Distributie notelor de pe IMDB pentru short films si feature films')

plt.show()
```

Se observa o distributie spre dreapta a categoriei 'Short Film', lucru care inseamna ca au o nota mai buna in medie pe IMDB, fata de 'Feature films'

**Care sunt genurile de filme produse cel mai frecvent de Netflix?**

```python
topgenre = netflix['Genre'].value_counts().nlargest(10)
```

```python
topgenre
```

```python
fig=plt.figure()
fig.add_axes([0,0,2,2])
topgenre.plot(kind='pie')
plt.title('Top 10 genuri cu cele mai numeroase filme produse de Netflix')
plt.axis('equal')
plt.show()
```

Productiile Netflix sunt compuse in mare parte din documentare, urmate de drame, comedii, comedii-romantice si thrillere.

**Cate filme au fost produse in fiecare an?(pana la 01/06/2021)** 

```python
def str_to_date(x):
    return datetime.strptime(x, '%Y-%m-%d')
```

```python
netflix['Premiere']=netflix['Premiere'].apply(lambda x: str_to_date(x))
```

```python
netflix['Year'] = netflix['Premiere'].dt.year
```

```python
netflix.head()
```

```python
plt.figure(figsize=(15,7), dpi=80)
sns.countplot(x='Year', data=netflix)
```
